```python
with open('../config.json') as fp:
    config_obj = json.load(fp)
config_obj
```

```python
group_id = 'flink-test-2'

price_src = KafkaSource.builder() \
    .set_bootstrap_servers(config_obj['kafka.brokers']) \
    .set_topics(config_obj['kafka.topic.price.name']) \
    .set_group_id(group_id) \
    .set_starting_offsets(KafkaOffsetsInitializer.earliest()) \
    .set_value_only_deserializer(SimpleStringSchema()) \
    .build()

ds = env.from_source(price_src, WatermarkStrategy.for_monotonous_timestamps(), 'price_src')

def hello(data):
    return data

ds.map(hello)

env.execute('Flink-kafka example').print()
```
